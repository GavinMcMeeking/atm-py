```python
from matplotlib.colors import LogNorm
def _getXYZ(dist):
    """ This will create three arrays, so when plotted with pcolor each pixel will represent the exact bin width"""
    binArray = np.repeat(np.array([dist.bins]),dist.data.index.shape[0], axis=0)
    timeArray = np.repeat(np.array([dist.data.index.values]), dist.bins.shape[0], axis = 0).transpose()
    ext = np.array([np.zeros(dist.data.index.values.shape)]).transpose()
    Z =  np.append(dist.data.values, ext, axis = 1)
    return timeArray,binArray,Z
def plot_distribution(dist):
    X,Y,Z = _getXYZ(dist)
    f,a = plt.subplots()
    pc = a.pcolormesh(X,Y,Z, norm = LogNorm())
    a.set_yscale('log')
    a.set_ylim((dist.bins[0],dist.bins[-1]))
    a.set_xlabel('Time (UTC)')
    a.set_ylabel('Diameter (nm)')
    f.colorbar(pc)
    return f,a,pc
```

```python
def zoom_time(dist, start = None, end = None):
    """'2014-11-24 16:02:30'"""
    dist.data = dist.data.truncate(before=start, after = end)
    return 
```

```python
def average_overTime(dist, window='1S'):
    window = window
    dist.data = dist.data.resample(window, closed='right',label='right')
    return
```

```python
def average_overAllTime(dist):
    singleHist = np.zeros(dist.data.shape[1])
    for i in xrange(dist.data.shape[1]):
        line = dist.data.values[:,i]
        singleHist[i] = np.average(line[~np.isnan(line)])
    return singleHist
```
