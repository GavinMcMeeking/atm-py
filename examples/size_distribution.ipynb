```python
from atmPy import sizedistribution
```

```python
%matplotlib inline
```

# simulate a sizedistribution

```python
sd = sizedistribution.simulate_sizedistribution(diameter=[0.01, 2.5],
                                                numberOfDiameters=100, 
                                                centerOfAerosolMode=0.2, 
                                                widthOfAerosolMode=0.2, 
                                                numberOfParticsInMode=1000)
sd.plot_distribution()
```

# convert to other reprisentations

```python
sd.convert2dNdlogDp()
sd.plot_distribution()
```
