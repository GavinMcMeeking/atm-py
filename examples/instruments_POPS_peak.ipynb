```python
def apply_calibration(m, calibrationInstance):
    m.data['Diameter'] = pd.Series(calibrationInstance.calibrationFunction(m.data.Amplitude.values), index = m.data.index)
    
    where_tooBig = np.where(m.data.Amplitude > cal.data.amp.max())
    where_tooSmall = np.where(m.data.Amplitude < cal.data.amp.min())
    tooSmall = len(where_tooSmall[0])
    tooBig = len(where_tooBig[0])
    m.data.Masked.values[where_tooBig] = 1
    m.data.Masked.values[where_tooSmall] = 1
    misc.msg('\t %s from %s peaks (%.1i %%) are outside the calibration range (amplitude = [%s, %s], diameter = [%s, %s])'%(tooSmall + tooBig, len(m.data.Amplitude),100 * float(tooSmall + tooBig)/float(len(m.data.Amplitude)) , cal.data.amp.min(),  cal.data.amp.max(), cal.data.d.min(), cal.data.d.max()))
    misc.msg('\t\t %s too small'%(tooSmall))
    misc.msg('\t\t %s too big'%(tooBig))
    return
```

```python
def plot_timeVindex(m):
    notMasked = np.where(m.data.Masked == 0)
    f,a = plt.subplots()
    g, = a.plot(m.data.index[notMasked],'o')
    a.set_title('Time as a function of particle index')
    a.set_ylabel("Time (UTC)")
    a.set_xlabel('Particle index')
    return f,a,g

def _plot_somethingVtime(m, what,title,ylabel):
    notMasked = np.where(m.data.Masked == 0)
    f,a = plt.subplots()
    g, = a.plot(m.data.index[notMasked], m.data[what].values[notMasked],'o')
    a.set_title(title)
    a.set_xlabel("Time (UTC)")
    a.set_ylabel(ylabel)
    return f,a,g

def plot_widthVtime(m):
    return _plot_somethingVtime(m,'Width','Peak width as a function of time','Width (sampling steps)')


def plot_diameterVtime(m):
    return _plot_somethingVtime(m,'Diameter', 'Peak diameter as a function of time','Diameter (nm)')


def plot_amplitudeVtime(m):
    return _plot_somethingVtime(m,'Amplitude','Peak amplitude as a function of time', 'Amplitude (digitizer bins)' )
```

```python
def get_countRate(m,average = None):
    """average: string, e.g. "5S" for 5 seconds"""
    notMasked = np.where(m.data.Masked == 0)
    unique = np.unique(m.data.index.values[notMasked])
    numbers = np.zeros(unique.shape)
    deltaT = np.zeros(unique.shape)
    countsPerSec = np.zeros(unique.shape)
    dt = unique[1]-unique[0]
    for e,i in enumerate(unique):
        if e:
            dt = unique[e] - unique[e-1]
        numbers[e] =  float(len(np.where(m.data.index.values[notMasked] == i)[0]))
        deltaT[e] = dt / np.timedelta64(1, 's') # delta in seconds
        countsPerSec[e] = numbers[e] / deltaT[e]

    countRate = pd.DataFrame(np.array([numbers,deltaT,countsPerSec]).transpose(), index = unique, columns=['No_of_particles', 'DeltaT_s', 'CountRate_s'])    
    if average:
        countRate = countRate.resample(average,closed = 'right', label='center')
    return countRate
```

```python
defaultBins = np.array([0.15, 0.168, 0.188, 0.211, 0.236, 0.264, 0.296, 0.332, 0.371, 0.416, 0.466, 0.522, 0.584, 0.655, 0.864, 1.14, 1.505, 1.987, 2.623, 3.462])
defaultBins *= 1000


def _peak2Distribution(peakInstance, bins=defaultBins, distributionType = 'number', differentialStyle = False):
    """Returns the particle size distribution normalized in various ways
    distributionType
        \t number:\t numbers only $\mu m^{-1}\, cm^{-3}$
        \t surface:\t surface area distribution, unit: $\mu m\, cm^{-3}$
        \t volume:\t  volume distribution, unit: $\mu m^{2}\, cm^{-3}$
    differentialStyle:\t     if False a raw histogram will be created, else:
        \t dNdDp: \t      distribution normalized to the bin width, bincenters are given by (Dn+Dn+1)/2
        \t dNdlogDp:\t    distribution normalized to the log of the bin width, bincenters are given by 10**((logDn+logDn+1)/2)

    """
    notMasked = np.where(peakInstance.data.Masked == 0)
    unique = np.unique(peakInstance.data.index.values[notMasked])
    N = np.zeros((unique.shape[0],bins.shape[0]-1))


    for e,i in enumerate(unique):
        condi = np.where(np.logical_and(peakInstance.data.Masked == 0, peakInstance.data.index.values == i))
        process = peakInstance.data.Diameter.values[condi]
        n,edg = np.histogram(process, bins = bins)
    #     print bins
    #     print process
    #     print N
    #     break
        N[e] = n

#     N,edg = np.histogram(peakInstance.data.Diameter.values[notMasked], bins = bins)
    N = N.astype(np.float)

    deltaT = (unique[1:]-unique[:-1]) / np.timedelta64(1,'s')
    deltaT = np.append(deltaT[0],deltaT)
    deltaT = np.repeat(np.array([deltaT]),bins.shape[0]-1, axis=0)
    N/=deltaT.transpose()
    
    bincenter = (edg[:-1] + edg[1:])/2.
    binwidth = edg[1:] - edg[:-1]

    if not differentialStyle:
        pass

    elif differentialStyle == 'dNdDp':
        N = N/binwidth

    elif differentialStyle == 'dNdlogDp':
        bincenter = 10**((np.log10(e[:-1]) + np.log10(e[1:]))/2.)
        binwidth = np.log10(e[1:]) - np.log10(e[:-1])
        N = N/binwidth
    else:
        raise ValueError('wrong type for argument "differentialStyle"')      

    if distributionType == 'surface':
        N *= (bincenter**2 * np.pi)
    elif distributionType == 'volume':
        N *= (bincenter**3 * np.pi/6.)
    elif distributionType == 'number':
        pass
    else:
        raise ValueError('wrong type for argument "distributionType"')    

    binstr = bins.astype(int).astype(str)
    cols=[]
    for e,i in enumerate(binstr[:-1]):
        cols.append(i+'-'+binstr[e+1])
    dataFrame = pd.DataFrame(N, columns=cols, index = unique)
    return sizedistribution.aerosolSizeDistribution(dataFrame,bins,bincenter, binwidth, distributionType, differentialStyle )
```
