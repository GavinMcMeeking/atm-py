## helpers

```python
def loadDistSave(fname,cal):
    m = peaks.read_PeakFile_Binary(fname)
    m.apply_calibration(cal)
    bins = np.logspace(np.log10(120),np.log10(3000), 200)
    dist = m.peak2numberdistribution(bins = bins)
    dist = dist.average_overTime()
    distfname = fname+'_distribution.csv'
    dist.save_csv(distfname)
    dist.plot_distribution()
    return dist
```

```python
def DistZoomPlot(dist,start = '2015-01-13 19:28:00', end = '2015-01-13 19:38:30'):
    dist_zoom = dist.zoom_time(start = start,end = end)
    dist_zoom.plot_distribution()
    averagAll = dist_zoom.average_overAllTime()
    f,a = plt.subplots()
    a.plot(dist_zoom.bincenters,averagAll)
    x = dist_zoom.bincenters.copy()
    a.semilogx()
    return x,averagAll
```
